```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
res = classifier("I love to learn.")
print (res)
```

We can spicify a model

```python
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
res = classifier("In Hamburg city the weather is not good.")
print (res)
```

```python
print (classifier ("Whatever I can, I will help people."))
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

res = classifier("I like pizza more than very spicy Biryani.")
print (res)

model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained (model_name)
tokenizer = AutoTokenizer.from_pretrained (model_name)

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

res = classifier("I like pizza more than very spicy Biryani.")
print (res)
```

How Tokenizer works?

```python
sequence = "Using a Transformer network is simple"
res = tokenizer (sequence)
print (res)
#same...
tokens = tokenizer. tokenize (sequence)
print (tokens)
ids = tokenizer.convert_tokens_to_ids (tokens)
print (ids)
decoded_string = tokenizer.decode(ids)
print (decoded_string)
```

```python
from transformers import pipeline
# classifier = pipeline('zero-shot-classification', model = "niting3c/llama-2-7b-hf-zero-shot-prompt").  # due to largre size and limited resources of colab free v
classifier = pipeline('zero-shot-classification', model = "facebook/bart-large-mnli")
```

Zero shot classification:

```python
text_to_classify = "I love to visit all the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(text_to_classify, candidate_labels)
```

topic modelinng:

```python
text_to_classify = "Dance is an art form, often classified as a sport, consisting of sequences of body movements with aesthetic and often symbolic value, either improvised or purposefully selected."
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(text_to_classify, candidate_labels)
```

Multilabling:

```python
text_to_classify = "The Internet is the global system of interconnected computer networks that uses the Internet protocol suite (TCP/IP)[b] to communicate between networks and devices. It is a network of networks that consists of private, public, academic, business, and government networks of local to global scope, linked by a broad array of electronic, wireless, and optical networking technologies. The Internet carries a vast range of information resources and services, such as the interlinked hypertext documents and applications of the World Wide Web (WWW), electronic mail, telephony, and file sharing."
candidate_labels = ['internet', 'art', 'technology', 'dance']
classifier(text_to_classify, candidate_labels, multi_label=True)

```
